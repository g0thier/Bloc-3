```python
dataset = dataset.dropna(subset=['CPI'])
```

```python
pourcentage_valeur_manquante = 100*dataset.isnull().sum()/dataset.shape[0]
print(pourcentage_valeur_manquante)
```

```python
dataset['Unemployment'] = dataset['Unemployment'].fillna(dataset['Unemployment'].mean())
```

```python
pourcentage_valeur_manquante = 100*dataset.isnull().sum()/dataset.shape[0]
print(pourcentage_valeur_manquante)
```

```python
dataset = dataset.sort_values(by= ['Date'])
```

```python
dataset.head()
```

```python
dataset.reset_index(inplace=True, drop=True)

for index in range(len(dataset)):
    if np.isnan(dataset['Temperature'][index]):
        dataset['Temperature'][index] = (dataset['Temperature'][index-1] + dataset['Temperature'][index+1] )/2

for index in range(len(dataset)):
    if np.isnan(dataset['Fuel_Price'][index]):
        dataset['Fuel_Price'][index] = (dataset['Fuel_Price'][index-1] + dataset['Fuel_Price'][index+1] )/2
```

```python
pourcentage_valeur_manquante = 100*dataset.isnull().sum()/dataset.shape[0]
print(pourcentage_valeur_manquante)
```
